```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
import seaborn as sns
```

```python
wine = pd.read_csv("/content/wine.csv")
wine.head()
```

# **PCA**

```python
pca = PCA()
pca_values = pca.fit_transform(wine_norm)
```

```python
pca_values
```

```python
pca = PCA(n_components = 10)
pca_values = pca.fit_transform(wine_norm)
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,decimals =4 )*100)
var1
```

```python
pca.components_
```

```python
wine.columns
```

```python
plt.plot(var1, color="red")
```

```python
pca_values[:,:4]
```

```python
#Plot between PCA1 and PCA2'
x = pca_values[:,0:1]
y = pca_values[:,1:2]
plt.scatter(x,y)
```

```python
final_df=pd.concat([wine['Type'],pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
sns.scatterplot(data=final_df, x='PC1', y='PC2', hue='Type')
```

# **BY Using Hierarchical Clustering**

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
dendrogram = sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hc
```

```python
y=pd.DataFrame(hc.fit_predict(wine_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
wine2=wine.copy()
wine2['clustersid']=hc.labels_
wine2
```

# **Using K-Mean Clustering**

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS');
```

```python
clusters3=KMeans(3,random_state=30).fit(wine_norm)
clusters3
```

```python
clusters3.labels_
```

```python
wine3=wine.copy()
wine3['clusters3id']=clusters3.labels_
wine3
```

```python
wine3['clusters3id'].value_counts()
```
